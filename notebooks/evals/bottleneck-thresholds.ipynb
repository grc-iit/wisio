```python
%autoreload 2

from wisio.recorder import RecorderAnalyzer

# log_dir = "/p/vast1/iopp/recorder_app_logs/cm1/nodes-32/workflow-4/_parquet"
log_dir = "/p/vast1/iopp/recorder_app_logs/montage_pegasus/nodes-32/_parquet"

recorder_analyzer = RecorderAnalyzer(
    cluster_manager_args=dict(
        force_local=True,
        cluster_settings=dict(
            cores=4,
            dashboard_port=3646,
            local_directory="/var/tmp/dask-recorder-mpv",
            log_file="%J.log",
            worker_queue="pdebug"
        )
    ),
    working_dir='.recorder',
    debug=True
) 

recorder_analyzer
```

```python
def print_scores(bottlenecks):
    scores = None

    for view_key in bottlenecks.keys():
        # print(view_key)
        score_values = bottlenecks[view_key]['high_level_view']['duration_score'].value_counts()
        # print(score_values.compute())
        if scores is None:
            scores = score_values.compute()
        else:
            scores = scores.add(score_values.compute(), fill_value=0)

    print(scores)
```

```python
%%timeit -r 10 -n 1

res = recorder_analyzer.analyze_parquet(log_dir=log_dir, checkpoint=False, cutoff=0.01)

main_view, views, bottlenecks = res

print_scores(bottlenecks)
```
