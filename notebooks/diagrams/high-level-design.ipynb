```python
from diagrams import Cluster, Diagram
from diagrams.programming.flowchart import Action, Decision, Document, MultipleDocuments
from IPython.display import display, Image

with Diagram('WisIO', show=False, filename='wisio'):
    inputs = MultipleDocuments('Recorder I/O Traces')
    outputs = MultipleDocuments('I/O Bottlenecks')

    with Cluster('WisIO'):
        with Cluster('Converter Tool', direction='TB'):
            converter_0 = Action('Convert I/O traces \n into Parquet')
            converter_1 = Action('Apply \n transformations')
            converters = [
                converter_0,
                converter_1,
            ]
            # converter_0 >> converter_1 >> converter_2

        with Cluster('Multifaceted Views', direction='TB'):
            view_0 = Document('Time-based')
            view_1 = Document('File-based')
            view_2 = Document('Process-based')
            views = [
                view_0,
                view_1,
                view_2
            ]

        with Cluster('Bottleneck Generator'):
            gen_0 = Decision('Select significant \n I/O bottlenecks')
            gen_1 = Action('Generate observations')
            with Cluster('Rule Engine'):
                rule_0 = Action('Find reasons for \n observations')
                rule_1 = Decision('Produce diagnoses')

    inputs >> converter_0 >> converter_1 >> views >> gen_0 >> gen_1 >> rule_0 >> rule_1 >> outputs

display(Image(filename='wisio.png'))
```
